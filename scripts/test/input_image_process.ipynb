# deal with the compressed depth image from rosbag

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# read image
depth_raw = np.load('depth_raw_rs.npy')
fig = plt.figure()
plt.imshow(depth_raw, cmap='gray')
```

```python
# rescale image to 100 80
image_small = cv2.resize(depth_raw, (100, 80), interpolation = cv2.INTER_NEAREST)
```

```python
# get depth image in meter
MAX_DEPTH_METER = 10

image_small_meter = image_small / 1000  # transter depth from mm to meter
image_small_meter_ori = np.copy(image_small_meter)
image_small_meter[image_small_meter == 0] = MAX_DEPTH_METER
```

```python
# get depth image in gray (0-255)
image_gray = image_small_meter / MAX_DEPTH_METER * 255
image_gray_int = image_gray.astype(np.uint8)
# cv2.imshow('image_gray_int', image_gray_int)
```

```python

fig = plt.figure()
obs = 255 - image_gray_int
plt.imshow(image_gray_int, cmap='gray')
```
